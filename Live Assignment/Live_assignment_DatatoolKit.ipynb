17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns
```

```python

x = np.random.randn(100)
y = np.random.randn(100)

def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Q1'
    elif x < 0 and y > 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    elif x > 0 and y < 0:
        return 'Q4'

df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: determine_quadrant(row['x'], row['y']), axis=1)

plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='bright')

scatter_plot.set_title('Quadrant-wise Scatter Plot')
scatter_plot.set_xlabel('X-axis')
scatter_plot.set_ylabel('Y-axis')
plt.legend(title='Quadrant')

plt.show()
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.


```python

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))
data = pd.DataFrame({'categories': categories, 'values': values})
source = ColumnDataSource(data)
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Category', y_axis_label='Value', height=400, width=600)
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color=factor_cmap('categories', palette='Spectral5', factors=categories))
show(p)
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'.


```python
import plotly.graph_objs as go

np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.random.randn(100).cumsum()


fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis Label',
    yaxis_title='Y-axis Label'
)
fig.show()
```
